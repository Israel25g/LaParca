# Bloque de entradas

```python
promJulio = libro['Julio'].mean()
promJulio = promJulio * 100
promJulio = round(promJulio, 2) # Redondear a 2 decimales
print (promJulio)

# promJulio.to_csv('promedio.json', orient='records')
```

```python
otif = pd.pivot_table(libro, index='Cliente', values=['Agosto'], aggfunc='sum').reset_index()
otif = pd.DataFrame(otif)
otif['OTIF'] = otif['Agosto'] * 100
```

```python
otif_ago=otif[['Cliente','OTIF']]
prom_ago = otif_ago['OTIF'].mean()
print(prom_ago)
```
